```python
def stats(data):
    x = np.arange(len(data))  # Generate x values as indices (0, 1, 2, ...)
    slope = np.polyfit(x, data, 1)[0]

    plt.plot(x, data, 'o', label='Data Points')
    plt.plot(x, np.polyval(np.polyfit(x, data, 1), x), label='Regression Line')
    plt.xlabel('Index')
    plt.ylabel('Data')
    plt.title('Data Points with Regression Line')
    plt.legend()
    plt.grid(True)
    plt.show()

    x = np.arange(len(data))  # Generate x values as indices (0, 1, 2, ...)
    slope = np.polyfit(x, data, 1)[0]
    mean_value = statistics.mean(data) 
    print(slope / mean_value)

    print(statistics.stdev(data) / statistics.mean(data))
        
```

```python
import requests
from lxml.html import fromstring
def get_proxies():
    url = 'https://free-proxy-list.net/'
    response = requests.get(url)
    parser = fromstring(response.text)
    proxies = set()
    for i in parser.xpath('//tbody/tr')[:10]:
        if i.xpath('.//td[7][contains(text(),"yes")]'):
            #Grabbing IP and corresponding PORT
            proxy = ":".join([i.xpath('.//td[1]/text()')[0], i.xpath('.//td[2]/text()')[0]])
            proxies.add(proxy)
    return proxies
```

```python
import re
import requests
from bs4 import BeautifulSoup


# finds id of the product
def get_id(s):
    id = None
    for script in s.find_all('script'):
        id_regex = re.search('Market_LoadOrderSpread\(([ 0-9]+)\)', script.text)
        if id_regex:
            id = id_regex.groups()[0].strip()
            break
    return id

def get_order_price():
    name_url = "https://steamcommunity.com/market/listings/730/P250%20%7C%20Red%20Rock%20%28Battle-Scarred%29"
    html = requests.get(name_url).text
    soup = BeautifulSoup(html, 'lxml')
    id = get_id(soup)

    if id:
        id_url = f"https://steamcommunity.com/market/itemordershistogram?country=US&language=english&currency=1&item_nameid={id}&two_factor=0"
        html = requests.get(id_url).json()
        soup = BeautifulSoup(html['buy_order_summary'], 'lxml')

        not_format = soup.select_one('span:last-child').text
        formated_price = float(not_format[1:])
        return formated_price

    else:
        print("Could not get ID")
        exit()
get_order_price()
```

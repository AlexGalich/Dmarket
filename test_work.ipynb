```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
def stats(data):
    x = np.arange(len(data))  # Generate x values as indices (0, 1, 2, ...)
    slope = np.polyfit(x, data, 1)[0]

    plt.plot(x, data, 'o', label='Data Points')
    plt.plot(x, np.polyval(np.polyfit(x, data, 1), x), label='Regression Line')
    plt.xlabel('Index')
    plt.ylabel('Data')
    plt.title('Data Points with Regression Line')
    plt.legend()
    plt.grid(True)
    plt.show()

    x = np.arange(len(data))  # Generate x values as indices (0, 1, 2, ...)
    slope = np.polyfit(x, data, 1)[0]
    mean_value = statistics.mean(data) 
    print(slope / mean_value)

    print(statistics.stdev(data) / statistics.mean(data))
        
```

```python
data = [0.39, 0.39, 0.38, 0.38, 0.36, 0.36, 0.36, 0.36, 0.36, 0.36, 0.36, 0.36, 0.36, 0.36, 0.35]
stats(data)
```
